1. $\theta = 2$일 때 $0 \le x \le 20$ 구간을 1000 등분한 점에서 함수 $f(x) = 1 - e^{-\theta x}$의 값을 계산하여 출력하시오.

```python
# 1. 우선 x의 값을 만들기 위해서 다음을 수행한다. 
# 문제에서 주어지는 상수는 변수로 저장해서 이용한다. 
theta = 2

seq = list(range(0, 1001)) # x가 가질 수 있는 값은 1000등분 하기 때문에 1001개의 값을 가진다. 

x = []
for i in seq:
    x.append(i/50) # 0부터 1000까지 1간격의 정수 값을 0에서 20 사이의 값으로 변환하기 위해서 50으로 나누어 준다. 1000 / 20 = 50

print(x)
print(len(x))
print(x[0], x[-1])
```

```python
import math

fx = []

for x0 in x: # 리스트 자료 x에서 순서대로 하나씩 x0로 넘겨주고 반복문을 실행한다.
    fx0 = 1 - math.exp(-theta * x0) # 중간 과정의 결과를 fx0에 저장한다. 
    fx.append(fx0) # 계산된 fx0를 fx 리스트에 누적한다. 
    #fx.append(1 - math.exp(-theta*value)) # 한줄로 만들 수 있다.

print(fx[0], fx[-1])
print(len(fx))
```

2. $n=20$, $p=0.3$일 때 $x=0, 1, \dots, n$에서 함수 $f(x) = \binom{n}{x} p^x (1-p)^{n-x}$의 값을 계산하여 출력하시오.

```python
import math

n = 20 
p = 0.5

fx = []

x = list(range(n))

for x0 in x: #list(range(n)):
    fx0 = math.factorial(n) / (math.factorial(x0) * math.factorial(n-x0)) * math.pow(p, x0)*math.pow(1-p, n - x0)
    fx.append(fx0)

print(len(fx))
print(fx[:3], fx[-3])
```

5. 구간 $[-5, 5]$을 100 등분하는 점에서 함수 $f(x) = \left\{
     \begin{array}{ll}
       x & \text{if } x \gt 0 \\
       0 & \text{if} x \le 0
     \end{array}
   \right.$를 계산하시오.

```python
# [-5, 5]의 구간을 100 등분 하면 1001개의 값을 가진다.
# range 함수를 이용해서 -50에서 50까지 1단위 정수를 만들면 1001개의 값을 가진다.
seq = list(range(-50, 51)) # range는 끝값을 포함하지 않음

x = []

for i in seq:
    x.append(i/10) # 10으로 나누어 주면 -50에서 50까지의 정수는 -5에서 5까지의 실수로 변환된다.

fx = []

for x0 in x:
    if x0 > 0:
        fx.append(x0)
    else:
        fx.append(0)

print(len(fx))
print(fx[:5], fx[-5:])
```

10. random 모듈의 난수발생함수를 이용하여 10 명의 키와 체중이 저장된 리스트 heights, weights를 생성한 다음 10 명의 bmi가 저장된 리스트 bmis를 만들고 출력하시오.

```python
import random

heights = []
weights = []
for i in range(10): # 10번 반복하기 위해서 range 함수를 이용해서 0에서 9까지 정수를 발생시키면, 10번에 반복되어 진다. 
    heights.append(random.uniform(140, 200)) # 키가 가질 수 있는 값의 범위를 고려해야 한다. 
    weights.append(random.uniform(40, 100))  # 몸무게가 가질 수 있는 값의 범위를 고려해야 한다. 

print(heights)
print(weights)
```

```python
import random

heights = []
weights = []
while len(heights) <= 10:
    heights.append(random.uniform(140, 200))
    weights.append(random.uniform(40, 100))

print(len(heights))
print(heights)
print(weights)
```

```python
import random

heights = []
weights = []

i = 1
while i <= 10:
    heights.append(random.uniform(140, 200))
    weights.append(random.uniform(40, 100))
    #i=+1
    i = i + 1
    
print(len(heights))
print(heights)
print(weights)
```

```python
import math

bmis = []

for idx, value in enumerate(heights):
    bmi = weights[idx] / math.pow((value/100), 2)
    bmis.append(bmi)

print(bmis)
```
