* 변수의 id를 알아보는 예제 프로그램을 작성하시오.

**Note** 파이썬의 모든 객체는 고유한 ID(identifier)를 갖는다. ID는 정수이다.

**Note** 객체는 메모리에 존재하며 고유한 id를 갖는다. 어떤 객체를 참조한다는 것은 그 객체의 id를 저장하고 있다는 것이며 이 id를 이용해서 해당 객체에 접근한다.

```python
x = 100          # 정수 100이 객체이며, x는 이 정수 객체의 참조이다.
print(id(x))     # id(x)는 x가 참조하는 객체의 id를 반환한다.
s = 'github.com' # 문자열 `github.com'이 객체이며 s는 문자열 객체의 참조한다.
print(id(s))     # id(s)는 s가 참조하는 객체의 id를 반환한다.
y = x            # y는 x가 참조하는 객체를 참조한다.
                 # x와 y는 동일한 객체를 참조한다. 
print(id(y))     # id(y)는 y가 참조하는 객체의 id를 반환한다.
```

* 복소수를 만들고 실수부와 허수부를 출력하는 프로그램을 작성하시오.

```python
cn = complex(2, 3)
print("복소수: ", cn)
print("실수부: ", cn.real)
print("허수부: ", cn.imag)
```

* 복소수 간의 사칙연산을 보여주는 예제 프로그램을 작성하시오.

```python
a = complex(4, 3)
b = complex(2, -6)
print("덧셈: ", a + b)
print("뺄셈: ", a - b)
print("곱셈: ", a * b)
print("나눗셈: ", a / b)
```

* 복소수의 길이(절대값)를 계산하는 예제 프로그램을 작성하시오.

* 실수의 절대값을 계산하는 예제 프로그램을 작성하시오.

**Note** 내장함수 `abs()`와 `math` 모듈의 `fabs()` 함수는 모두 절대값을 구해주는 함수이다. `abs()` 함수는 절대값을 인자의 자료형으로 반환하고, `fabs()` 함수는 절대값을 항상 실수로 반환한다는 점이 다르다. 

```python
import math

print(math.fabs(-2.1))
print(math.fabs(10.1))
print(math.fabs(0.0))
```
