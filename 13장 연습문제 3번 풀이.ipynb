다음은 **심장병자료**를 이름이 country, wine, death인 리스트에 저장하는 코드이다.

```python
country = ["Australia", "Austria", "Belgium/Luxembourg", "Canada", "Denmark", 
           "Finland", "France", "Iceland", "Ireland", "Italy",
           "Netherlands", "New Zealand", "Norway", "Spain", "Sweden",
           "Switzerland", "United Kingdom", "United States", "West Germany"]
wine = [2.5, 3.9, 2.9, 2.4, 2.9, 0.8, 9.1, 0.8, 0.7, 7.9, 1.8, 1.9, 0.8, 6.5, 1.6, 5.8, 1.3, 1.2, 2.7]
death = [211, 167, 131, 191, 220, 297, 71, 211, 300, 107, 167, 266, 227, 86, 207, 115, 285, 199, 172]
```

* 각 list 객체에 저장된 값의 개수를 확인하시오.

```python
print(len(country))
print(len(wine))
print(len(death))
```

* wine, death에 저장된 값의 합, 평균, 편차, 편차 제곱합, 표준편차를 함수를 사용하지 않고 계산하시오.

```python
wine_sum = 0 # wine 합을 저장할 변수 이름
for w in wine:
    wine_sum += w #누적합

wine_n = len(wine) # wine의 자료 수
wine_mean = wine_sum / wine_n # wine 평균을 저장할 변수 이름

wine_dev = [] # wine 편차을 저장할 자료 이름
for w in wine:
    wine_dev.append(w - wine_mean)

import math 
wine_dev_sse = 0 # wine 편차제곱합을 저장할 변수 이름
for dev in wine_dev:
    wine_dev_sse += math.pow(dev, 2)

wine_std = math.sqrt(wine_dev_sse / (wine_n-1)) # wine 표준편차를 저장할 변수 이름, 표본 표준편차이므로 n-1의 자유도로 나눈다.

print('wine의 합 : ', wine_sum)
print('wine의 평균 : ', wine_mean)
print('wine의 편차제곱합 : ', wine_dev_sse)
print('wine의 표준편차 : ', wine_std)
```

```python
death_sum = 0 # death 합을 저장할 변수 이름
for w in death:
    death_sum += w #누적합

death_n = len(death) # death의 자료 수
death_mean = death_sum / death_n # death 평균을 저장할 변수 이름

death_dev = [] # death 편차을 저장할 자료 이름
for w in death:
    death_dev.append(w - death_mean)

import math 
death_dev_sse = 0 # death 편차제곱합을 저장할 변수 이름
for dev in death_dev:
    death_dev_sse += math.pow(dev, 2)

death_std = math.sqrt(death_dev_sse / (death_n-1)) # death 표준편차를 저장할 변수 이름, 표본 표준편차이므로 n-1의 자유도로 나눈다.

print('death의 합 : ', death_sum)
print('death의 평균 : ', death_mean)
print('death의 편차제곱합 : ', death_dev_sse)
print('death의 표준편차 : ', death_std)
```

* wine, death에 저장된 값의 합, 평균, 분산, 표준편차를 함수를 사용하여 계산하시오.

```python
import math
import statistics as stat

wine_sum1 = sum(wine) #합
wine_mean1 = stat.mean(wine) #평균
wine_var1 = stat.variance(wine) #분산
wine_std1 = stat.stdev(wine) #표준편차

print('wine의 합 : ', wine_sum1)
print('wine의 평균 : ', wine_mean1)
print('wine의 분산 : ', wine_var1)
print('wine의 표준편차 : ', wine_std)
```

```python
death_sum1 = sum(death) #합
death_mean1 = stat.mean(death) #평균
death_var1 = stat.variance(death) #분산
death_std1 = stat.stdev(death) #표준편차

print('death의 합 : ', death_sum1)
print('death의 평균 : ', death_mean1)
print('death의 분산 : ', death_var1)
print('death의 표준편차 : ', death_std1)
```

* wine, death에서 구한 편차의 합과 평균을 구하시오.

```python
wine_dev_sum = sum(wine_dev)
wine_dev_mean = stat.mean(wine_dev)

print(wine_dev_sum, wine_dev_mean)

death_dev_sum = sum(death_dev)
death_dev_mean = stat.mean(death_dev)

print(death_dev_sum, death_dev_mean)
```

* 서로 대응하는 wine과 death의 편차를 곱한 것의 합과 평균을 계산하시오.

```python
cov_seq = []
for idx, value in enumerate(wine_dev):
    cov_seq.append(value * death_dev[idx])

cov_sum = sum(cov_seq)    
cov_mean = stat.mean(cov_seq)

print(cov_sum)
print(cov_mean)
```

```python
cov_seq = []
for idx in range(len(wine_dev)):
    cov_seq.append(wine_dev[idx] * death_dev[idx])

cov_sum = sum(cov_seq)    
cov_mean = stat.mean(cov_seq)

print(cov_sum)
print(cov_mean)
```
