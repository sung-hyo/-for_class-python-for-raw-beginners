# 자료구조: 튜플(tuple)과 셋(set)

## 튜플

튜플(tuple)은 수학적 개념인 순서쌍을 구현한 자료구조이다.
튜플을 생성하는 방법은 `[]` 대신 `()`를 사용하여 생성한다는 것이 다르지만 리스트와 같이 인덱스를 이용해서 저장된 값에 접근하는 자료구조이다.
튜플과 리스트의 가장 큰 차이점은 리스트는 `mutable`하지만 튜플은 `immutable`하다는 것이다.
따라서 일단 생성된 튜플은 수정할 수 없다.
즉 튜플에 저장된 값을 수정하거나 추가하거나 삭제할 수 없다.
따라서 리스트 자료구조와 달리 `append()`, `remove()`, `insert()`, `clear()` 등과 같은 메소드가 없으며, `index()`, `count()` 등의 메소드는 있다.

````python
heam = ('Darwin Dawkins Park', 184, 76)
print(type(heam))
print(heam)
````

````python
for value in heam:
    print(value)
````

````python
for idx, value in enumerate(heam):
    print(idx, value)
````

````python
print(len(heam))
print(heam.count(184))
print(heam.index(184))
````

````python
heam[0] = 'Charles Darwin'
````

````python
heam.append(30)
````

리스트 자료구조와 같이 **튜플 간의 `+` 연산과 튜플에 대한 정수 곱셈 연산이 가능**하다.
이들 연산은 **피연산자인 튜플을 변경하는 것이 아니라 연산 결과를 새로운 튜플로 주기 때문에 가능하다.**
따라서 이들 연산 결과를 저장해야 다른 연산에 사용할 수 있다.

````python
print(heam + heam)
print(heam * 3)
print(heam)
````

**내장함수 `list()`, `tuple()`를 사용하면 리스트를 튜플로 또는 튜플을 리스트로 변환할 수 있다.**

````python
gndr = ['M']*2 + ['F']*4 + ['M'] + ['F']*2 + ['M'] + \
        ['F']*2 + ['M'] + ['F']*4 + ['M']*2
wgt = [62.0, 62.9, 36.1, 54.6, 48.5, 42.0, 47.4, \
          50.6, 42.0, 48.7, 40.3, 33.1, 51.9, 42.4, \
          34.5, 51.1, 41.2, 51.9, 46.9]
rate = [1792, 1666, 995, 1425, 1396, 1418, 1362, 1502,\
        1256, 1614, 1189, 913, 1460, 1124, 1052, 1347,\
        1204, 1867, 1439]
````

````python
print(type(gndr))
gndr = tuple(gndr)
print(type(gndr))
gndr = list(gndr)
print(type(gndr))
````

### 패킹(packing)

**여러 값을 한 변수에 저장하면 해당 변수는 자동적으로 모든 값을 저장한 튜플이 된다.**
이런 연산을 **패킹**이라 한다.

````python
pack = 'joongyang' + ' ' + 'park', 'GNU', 12345, 3.141592
print(type(pack))
print(pack)
````

튜플에 unpacking을 적용할 수 있다.
이때 2개 이상의 원소를 저장해야 할 경우에는 리스트가 이용된다.

````python
elem1, *elems = pack
print(elem1, elems)
````

튜플과 리스트 중 어떤 자료구조를 이용하는 것이 좋은가는 문제와 상황에 달려 있다.
기본적으로 자료구조에 저장된 값들이 변경해야 할 경우가 있느냐 없느냐에 달려 있다.

## 셋

값을 저장한 순서가 중요한 리스트나 튜플과 달리 **셋**(set)은 **값의 순서는 아무런 의미가 없고 값의 중복도 허용하지 않는 자료구조**이다.
셋은 수학적 개념인 **집합을 구현한 자료구조**라고 할 수 있다.
값이 저장된 순서에 아무 의미도 없기 때문에 리스트와 튜플에서 사용했던 **인덱스와 `[]`연산자를 셋에서는 사용할 수 없다.**
셋은 `mutable`하기 때문에 새로운 원소를 추가/삭제할 수 있으며, **특정한 값이 집합의 원소인가를 알아 보는 연산, 합집합 연산, 교집합 연산, 차집합 연산** 등을 수행하는 메소드와 연산자가 구현되어 있다.

**셋은 쉼표로 구분한 원소를 `{}`로 감싸서 생성한다.**
원소가 하나인 셋을 생성할 때도 값 뒤에 쉼표를 기재해야 한다.

````python
A = {'가', '나', '다'}  # brace, braket
print(A)
print(type(A))
````

**셋에 원소 하나를 추가할 때는 `add()` 메소드를 사용하고, 여러 개의 원소를 추가할 때는 추가할 원소의 리스트를 인자로 `update()` 메소드를 호출하면 된다.**
동일한 값을 여러 번 추가할 수는 있지만 셋은 중복을 허용하지 않기 때문에 한 번만 저장한다.

````python
A.add('라')
print(A)
A.update(['라', '라', '마', '마'])
print(A)
````

`list()` 함수와 `tuple()` 함수와 같이 `set()` 함수를 이용하면 리스트와 튜플을 셋으로 변환할 수 있다.

````python
temp = set(['라', '라', '마', '마'])
print(type(temp))
print(temp)
temp = tuple(temp)
print(type(temp))
print(temp)
temp = list(temp)
print(type(temp))
print(temp)
A.update(temp)
print(A)
````

특정한 값을 제거하고자 하면 제거할 값을 인자로 **`discard()`** 또는 **`remove()`** 메소드를 호출하면 된다.

````python
A.discard('가')
print(A)
A.remove('나')
print(A)
````

`discard()` 메소드와 `remove()` 메소드의 차이는 셋에 없는 값을 삭제할 때 나타난다.
`discard()` 메소드는 셋에 없는 값을 삭제하려고 하면 그냥 무시하지만 `remove()` 메소드는 `KeyError` 오류를 발생시킨다.

````python
A.discard('가')
print(A)
A.remove('가')
print(A)
````

**특정한 값이 집합의 원소인가를 알아볼 때는 `in` 연산자, 합집합 연산은 `|` 연산자 또는 `union()` 메소드, 교집합 연산은 `&` 연산자 또는 `intersection()` 메소드, 차집합 연산은 `-` 연산자 또는 `difference()` 메소드를 이용**하여 수행할 수 있다.

````python
print(A)
B = {'가', '나', '다'}
print(B)
````

````python
print('가' in A)
print('가' not in A)
````

````python
for ele in A:
    print(ele)
````

````python
for idx, ele in enumerate(A):
    print(idx, ele)
````

````python
AorB = A | B  # A | B, B | A
AcupB = A.union(B)
print(AorB)
print(AcupB)
````

````python
AandB = A & B  # A & B, B & A
AcapB = A.intersection(B)
print(AandB)
print(AcapB)
````

````python
AminusB = A - B   
AdiffB = A.difference(B)
print(AminusB)
print(AdiffB)
BminusA = B - A
BdiffA = B.difference(A)
print(BminusA)
print(BdiffA)
````

그 외에도 한 집합이 다른 집합의 **부분집합(subset)**인지 **확대집합(superset)**인지 또는 서로소 집합인지를 검사하여 결과를 논리값으로 반환하는 **`issubset()`**, **`issuperset()`**, **`isdisjoint()`** 메소드가 있다.

````python
ans = A.issubset(B)
print(ans)
````

````python
if B.issubset(A):
    print("ulala!!!")
    
if B.isdisjoint(A):
    print("hahaha")
````

### 튜플과 셋에 사용할 수 있는 내장함수

내장함수 `all()`, `any()`, `enumerate()`, `len()`, `max()`, `min()`, `sorted()`, `sum()`은 리스트 자료구조 뿐만 아니라 튜플과 셋 자료구조에도 사용할 수 있다.
